```python
from PIL import Image
import numpy
import requests
from io import BytesIO
```

```python
response = requests.get("https://www.cin.ufpe.br/~ajsilva/gato.jpg")
im = Image.open(BytesIO(response.content))
im
```

```python
tensor = numpy.array(im)
tensor.shape
```

```python
tensor[:, :, 2]
```

```python
tensor[:,:,0] *=0
tensor[:,:,1] *=0
im2 = Image.fromarray(tensor)
im2
```

```python
tensor2 = tensor[100:200, 200:400, :]
im_recorte = Image.fromarray(tensor2)
im_recorte
```

```python
tensor2.shape
```

```python
tensor3 = numpy.zeros((200, 100, 3), dtype='uint8')
im_nova = Image.fromarray(tensor3)
im_nova
```

```python
m = numpy.array([[0, -1], [1, 0]])
for x in range(100):
  for y in range(200):
    a = numpy.array([x, y])
    b = m @ a
    tensor3[b[0], b[1], 2] = tensor2[x, y, 2]
```

```python
im_nova = Image.fromarray(tensor3)
im_nova
```

```python
for i in range(100):
    tensor2[i, i, 1] = 255
```

```python
im_recorte = Image.fromarray(tensor2)
im_recorte
```
