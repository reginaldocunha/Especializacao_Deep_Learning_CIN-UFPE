```python
metadados = pd.read_json('http://dados.recife.pe.gov.br/dataset/2a9b1c39-0700-4ddf-9a10-b3c8d5d9396c/resource/8f25ed76-de28-4082-89a7-113cad9a31be/download/dengue-metadados2016.json')
```

```python
campos_json = metadados.loc['campos', 'metadados']
print("Objeto JSON dos campos:")
print(campos_json)
```

```python
# Criando um dicionário com listas vazias para cada coluna
campos_dict = {'codigo': [], 'descricao': [], 'tipo': [], 'tamanho': [], 'valores_permitidos': []}

# Iterando sobre os itens do JSON e preenchendo as listas no dicionário
for campo in campos_json:
    campos_dict['codigo'].append(campo['codigo'])
    campos_dict['descricao'].append(campo['descricao'])
    campos_dict['tipo'].append(campo['tipo'])
    campos_dict['tamanho'].append(campo['tamanho'])
    campos_dict['valores_permitidos'].append(campo['valores_permitidos'])

# Criando o DataFrame a partir do dicionário
campos_df = pd.DataFrame(campos_dict)

# Exibindo o DataFrame resultante
print("DataFrame dos campos:")
print(campos_df)
```
